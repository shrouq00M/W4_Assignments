```python

#Import packages

import pandas
```

```python
#Create the variables to store the parameters for a model

Revenue_growth_factor= 1.18 
EBITDA_margin= 0.45
Taxes= 0.21
capital_expenditures= 0.03 
working_capital_increase= 0.06
EBITDA_Exit_Multiple= 32.70
discount_Factor= 0.94

```

```python
#Create the index

dates = pandas.date_range(start='2022' , end = '2032' ,freq ='Y')
dates
```

```python
#Create a DataFrame

model_df = pandas.DataFrame(index=dates)
terminal_value_df = pandas.DataFrame()
model_df
```

```python
#Populate the DataFrame
#Calculate as columns:Revenue,EBITDA,Taxes,Capital Expenditures,Increase in Working Capital,Cash Flow,Discount Factor,Present Value of Future Cash Flow
#Revenue growth to estimate future sales

start_rev = 222

rev_projection = [start_rev * Revenue_growth_factor]

for i in range(1,len(dates)):
    rev_projection.append(rev_projection[i-1] * Revenue_growth_factor)
    
model_df['revenue']= rev_projection
   
```

```python
model_df[['revenue']]
```

```python
model_df['EBITDA']= model_df['revenue'] * EBITDA_margin
model_df
```

```python
model_df['Taxes']= model_df['EBITDA'] * -Taxes
model_df
```

```python
model_df['capital_expenditures']= model_df['revenue'] * - capital_expenditures
model_df
```

```python
model_df['working_capital_increase']= model_df['revenue'] * - working_capital_increase
model_df
```

```python
model_df['Cash_flow']= model_df.drop(columns=['revenue']).sum(axis=1)
model_df
```

```python
model_df = model_df.round(2)
model_df
```

```python
model_df.sum(axis=0)
```

```python
discount_Factors = [ discount_Factor ** i for i in range (1,len(dates)+1) ]
discount_Factors * model_df['Cash_flow'].values
```

```python
discount_Factors = [ discount_Factor ** 
                    i for i in range (1,len(dates)+1) ]
discount_table = pandas.DataFrame(index = discount_Factors)
discount_table['Present_Value_FCF']= discount_Factors * model_df['Cash_flow'].values
discount_table
```

```python
#Calculate the terminal value and total present value of Future Cash Flows

Present_Value_of_Projected_fcf = discount_table['Present_Value_FCF'].sum().round(2)

revenue_guidance_in_2032 = model_df['revenue'].iloc[-1]* Revenue_growth_factor

estimated_EBITDA = revenue_guidance_in_2032*EBITDA_margin

EBITDA_for_2030 = estimated_EBITDA * EBITDA_Exit_Multiple
```

```python
#print the required output

print(f" Our projected Present Value of Projected Future Cash Flows are:$ {Present_Value_of_Projected_fcf}")

print(f" We are providing revenue guidance in 2032 to be: ${revenue_guidance_in_2032.round(2)} and estimated EBITDA around:{estimated_EBITDA.round(2)}")

print(f" The terminal value at this time given a multiple of {EBITDA_Exit_Multiple} and the above EBITDA for 2030, would equal:{EBITDA_for_2030.round(2)}")

```
